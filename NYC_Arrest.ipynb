```python
##Import Packages 
import pandas as pd

csvs = ('split_file_part1.csv','split_file_part2.csv',
        'split_file_part3.csv','split_file_part4.csv',
        'split_file_part5.csv','split_file_part6.csv')

## To hold all DataFrames
all_dfs = []

for i in csvs:
    
    df = pd.read_csv(i)


## Data Transformation
    df['Category'] = df['LAW_CAT_CD'].replace({'F': 'Felony', 'M':  'Misdemeanor', 'V': 'Violation'})
    df['Arrest_Boro'] = df['ARREST_BORO'].replace({'B':'Bronx','S':'Staten Island','M':'Manhattan',
                                               'K':'Brooklyn','Q':'Queens'})
    df['Jurisdiction'] = df['JURISDICTION_CODE'].replace({0.0:'Patrol',1.0:'Transit',2.0:'Housing'})
    df['Jurisdiction'] = df['Jurisdiction'].where(df['JURISDICTION_CODE'] < 3.0, 'Non NYPD')
    df['Age'] = df['AGE_GROUP'].replace({'<18':'Under 18','18-24':'Young Adults','25-44':'Adults',
                                     '45-64':'Middle Age','65+':'Seniors'})
    df['Gender'] = df['PERP_SEX'].replace({'F':'Female','M':'Male'})


##Renaming COlumns
    df.rename(columns={'ARREST_KEY': 'Arrestid', 'ARREST_DATE': 'Adate','PD_CD':'Grancode','PD_DESC':'Grandesc',
                   'KY_CD':'Gencode','OFNS_DESC':'Gendesc','Arrest_Boro':'Aboro','ARREST_PRECINCT':'Aprecinct',
                   'PERP_RACE':'Race'}, inplace=True)


##Selection of required Columns
    df1 = df[['Arrestid', 'Adate', 'Grancode', 'Grandesc','Gencode','Gendesc',
       'Category', 'Aboro','Aprecinct', 'Jurisdiction', 'Age', 'Gender' ,'Race', 'Latitude', 'Longitude']]

## Drop rows with null values
    df_final = df1.dropna()
    all_dfs.append(df_final)
    
## Final DataFrame with all records

final_df = pd.concat(all_dfs, ignore_index=True)
final_df.to_csv('NYC_Arrest_Data.csv', index=False)
print("Success")
```
